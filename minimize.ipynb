```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize

def func(x):
    y = np.sin(x)* (x-1)**2 
    return y

x = np.linspace(-6, 6)
y = func(x)
y = np.array(y)


# initial guess : let's guess 5 according to the plot
x0 = 5.0

print(optimize.minimize(func, x0, method='Nelder-Mead'))

plt.figure(figsize=(6,4))
plt.plot(x, y, label = " $ f(x) = sin(x)*(x-1)^2 $")
plt.axvline(5.16046143, color = 'r', label = "Minimum")
plt.axhline(func(5.16046143), color = 'r')
plt.title('Defined function')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.grid()
plt.legend()
plt.savefig('Minimize.png')
plt.show()
```
