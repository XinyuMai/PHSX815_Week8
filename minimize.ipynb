```python
import numpy as np
from matplotlib import pyplot as plt
from scipy import optimize

def func(x):
    y = np.sin(x)* (x-1)**2 
    return y

x = np.linspace(-6, 6)
y = func(x)
y = np.array(y)


# initial guess : let's guess 5 according to plot
x0 = 5.0
result = optimize.minimize(func, x0, method='Nelder-Mead')
print(result)

plt.figure(figsize=(6,4))
plt.plot(x, y, label = "Function: $ f(x) = sin(x)*(x-1)^2 $")
plt.grid()
plt.axvline(5.16046143, color = 'r', label = "Minimum")
plt.title('Defined function')
plt.ylabel('f(x)')
plt.xlabel('x')
plt.legend()
plt.show()
```
